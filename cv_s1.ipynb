```python
from google.colab.patches import cv2_imshow
import cv2 
import numpy 
from skimage import io
```

```python
img1 = cv2.imread('cat.jpg')

cv2_imshow(img1)
```

```python
minn = numpy.amin(img1) #min value pixel in image
maxx = numpy.amax(img1) #max value pixel in image
print(minn)
print(maxx)
# if we want to see rgb image min max use max and min, for grayscale image we use amin and amax
```

```python
median = cv2.medianBlur(img1, 9)
cv2_imshow(median) #90% noise is added to our main image
```

```python
std = cv2.meanStdDev(img1)
print(std)
```
